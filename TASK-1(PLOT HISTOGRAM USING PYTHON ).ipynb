# plot a histogram in Python


Step 1: Import pandas and numpy

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

Step 2: Generate Rando Data

```python
mu = 168 #mean
sigma = 5 #stddev
sample = 250
np.random.seed(0)
height_f = np.random.normal(mu, sigma, sample).astype(int)

mu = 176 #mean
sigma = 6 #stddev
sample = 250
np.random.seed(1)
height_m = np.random.normal(mu, sigma, sample).astype(int)
```

```python
height_f
```

Step 3: Ready The Data

```python
gym = pd.DataFrame({'height_f': height_f, 'height_m': height_m})
```

```python
gym
```

### line chart

```python
gym.plot()
```

## Nicer line chart

```python
gym.groupby('height_m').count().plot()
```

# Bar charts

```python
gym.groupby('height_m').count().plot.bar()
```

```python
gym.groupby('height_f').count().plot.bar()
```

### Step 4: Plot the histogram in Python

```python
gym.hist()
```

```python
gym.hist(bins=20)
```

Plot the histograms on the  chart

```python
gym.plot.hist(bins=20)
```

```python
gym.plot.hist(bins=20, alpha=0.7)
```
